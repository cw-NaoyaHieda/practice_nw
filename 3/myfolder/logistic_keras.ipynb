モデル設定

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD

np.random.seed(0)  # 乱数シード
#層構造の設定 Dense(入力二次元、出力一次元、activationで活性化関数)
model = Sequential([
    # Dense(input_dim=2, output_dim=1),  # Keras 1
    Dense(input_dim=2, units=1),         # Keras 2
    Activation('sigmoid')
])
#model.addでどんどん層を追加していく書き方が可能　その場合
#model = Sequential()
#model.add(Dense(input.dim=2, units=1))
#model.add(Activation('sigmoid'))

#確率的勾配法(SGD)は下記の一行で実装 lrは学習率
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
```

モデル学習

```python
# ORゲート
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [1]])

#学習も一行
# model.fit(X, Y, nb_epoch=200, batch_size=1)  # Keras 1
model.fit(X, Y, epochs=200, batch_size=1)      # Keras 2
```

学習結果の確認

```python
#推定クラスと確率を同時に出力
classes = model.predict_classes(X, batch_size=1)
prob = model.predict_proba(X, batch_size=1)

print('classified:')
print(Y == classes)
print()
print('output probability:')
print(prob)
```
